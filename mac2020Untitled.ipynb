```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def plot_complex(complex_func):
    """ Domain Coloring """
    x, y = np.meshgrid(np.linspace(-4, 4, 400), np.linspace(-4, 4, 400))
    z = x + y*1j

    angles = (np.angle(complex_func(z)) + 2*np.pi) % (2*np.pi) # pix
    im = plt.imshow(angles, origin="lower", cmap="hsv") # origin="lower": y
    plt.axis('off') # 
    plt.colorbar(im, fraction=0.046, pad=0.04) # colorbar
    

plot_complex(np.sin)
```

```python

# GAUSSIAN_01
# https://python.atelierkobato.com/gaussian/

import numpy as np
import matplotlib.pyplot as plt

# 
def gauss(x, a = 1, mu = 0, sigma = 1):
    return a * np.exp(-(x - mu)**2 / (2*sigma**2))

# Figure
fig = plt.figure(figsize = (8, 6))

# FigureAxes
ax = fig.add_subplot(111)

# Axes'Gaussian Function'
ax.set_title("Gaussian Function", fontsize = 16)

# 
ax.grid()

# 
ax.set_xlabel("x", fontsize = 14)
ax.set_ylabel("y", fontsize = 14)

# 
ax.set_xlim([-4, 8])
ax.set_ylim([0, 1.2])

# -480.1
x = np.arange(-4, 8, 0.1)

# 
f1 = gauss(x)
f2 = gauss(x, a = 1, mu = 2, sigma = .2)

# Axes
ax.plot(x, f1, color = "red", label = "a=1.0, =0, =1")
ax.plot(x, f2, color = "blue", label = "a=0.5, =-4, =2")

# 
ax.legend(fontsize = 14)
```

```python
# GAUSSIAN_03

import numpy as np
import matplotlib.pyplot as plt

# 
def gauss(x, a = 1, mu = 0, sigma = 1):
    return a * np.exp(-(x - mu)**2 / (2*sigma**2))

# Figure
fig = plt.figure(figsize = (8, 6))

# FigureAxes
ax = fig.add_subplot(111)

# Axes'Gaussian Function'
ax.set_title("Gaussian Function", fontsize = 16)

# 
ax.grid()

# 
ax.set_xlabel("x", fontsize = 14)
ax.set_ylabel("y", fontsize = 14)

# 
ax.set_xlim([-3, 3])
ax.set_ylim([0, 3])

# -330.1
x = np.arange(-3, 3, 0.1)

# f
f = 0

# 
for k in range(5):
    g = gauss(x, mu = k)
    ax.plot(x, g, label = "{}".format(k))
    f += g

# Axes
ax.plot(x, f, color = "black", label = "0+1+2+3+4")

# 
ax.legend()

# 
plt.savefig("gauss.png", bbox_inches = "tight")
```

```python
from sympy import*
import numpy as np
from matplotlib import pyplot as plt
 
# 
cal_x = np.arange(-10.0, 10.1, 0.1)
cal_y = np.exp(cal_x)
 
# 
x = Symbol('x')
 
# 
sol = []                                                # 
a = 0                                                   # 
for i in range(6):
    taylor = series(exp(x), x=x, x0=a, n=i+2).removeO() # removeO()
    print(taylor)                                       # 
    taylor_y = lambdify(x, taylor, 'numpy')             # numpy
    sol.append(taylor_y(cal_x))                         # 
 
# ---------------------------------------------------
# 
plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'Times New Roman'
 
# 
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
 
# 
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')
 
# 
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_xlim(-2, 2)
ax1.set_ylim(-2, 8)
 
# 
ax1.plot(cal_x, cal_y, label='Theory', lw=3, color='black')
for j in range(len(sol)):
    ax1.plot(cal_x, sol[j], label='Series:n=' + str(j+2))
 
fig.tight_layout()
 
# 
plt.legend()
plt.show()
plt.close()
# ---------------------------------------------------------------------
```

```python
import numpy as np
from matplotlib import pyplot as plt
 
# 
cal_x = np.arange(-10.0, 10.1, 0.1)
cal_y = np.sin(cal_x)
 
# 
x = Symbol('x')
 
 
# 
sol = []                                                # 
a = 0                                                   # 
for i in range(6):
    taylor = series(sin(x), x=x, x0=a, n=2*(i+1)).removeO() # removeO()
    print(taylor)                                       # 
    taylor_y = lambdify(x, taylor, 'numpy')             # numpy
    sol.append(taylor_y(cal_x))                         # 
 
# ---------------------------------------------------
# 
plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'Times New Roman'
 
# 
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
 
# 
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')
 
# 
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_xlim(-3, 3)
ax1.set_ylim(-1.5, 1.5)
 
# 
ax1.plot(cal_x, cal_y, label='Theory', lw=3, color='black')
for j in range(len(sol)):
    ax1.plot(cal_x, sol[j], label='Series:n=' + str(j+2))
 
fig.tight_layout()
 
# 
plt.legend()
plt.show()
plt.close()
# ---------------------------------------------------------------------
```
